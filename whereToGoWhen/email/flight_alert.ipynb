```python
# Function to convert UTC time to a specified timezone
def convert_time(row, column_name, new_timezone):
    # Parse the datetime and set timezone to UTC
    dt = pd.to_datetime(row[column_name]).replace(tzinfo=pytz.UTC)

    # Convert to the new timezone
    return dt.astimezone(pytz.timezone(new_timezone))


def fetch_and_process_active_flights(df):
    
    # Filter active flights and create a copy to avoid SettingWithCopyWarning
    df_active = df[df['flight_status'] == 'active'].copy()
    df_scheduled = df[df['flight_status'] == 'scheduled'].copy()
    df_scheduled.to_csv('scheduled_flights.csv')
    # Convert departure times to local and New York time
    nyc_timezone = 'America/New_York'
    for col in ['departure.scheduled', 'departure.estimated']:
        df_active[f'{col}.local'] = df_active.apply(
            lambda row: convert_time(row, col, row['departure.timezone']), axis=1)
        df_active[f'{col}.NYC'] = df_active.apply(
            lambda row: convert_time(row, col, nyc_timezone), axis=1)

    # Select and sort relevant columns
    columns_to_display = [
        'flight.number', 'flight_date', 'flight_status', 'departure.airport',
        'arrival.airport','departure.timezone', 'departure.scheduled', 'departure.scheduled.local',
        'departure.scheduled.NYC', 'departure.estimated', 'departure.estimated.local',
        'departure.estimated.NYC'
    ]
    return df_active[columns_to_display].sort_values(by=['departure.scheduled.NYC'])
```

```python
import pandas as pd
import pytz

params = {
    'access_key': ''
}

api_result = requests.get('http://api.aviationstack.com/v1/flights', params)

json_data = api_result.json()
# Convert to DataFrame
df = pd.json_normalize(json_data['data'])

# Function to convert UTC datetime strings to EST
def convert_to_est(dt_str):
    if pd.isna(dt_str):
        return None  # Return None if data is missing

    # Parse the datetime string (assuming it's in UTC)
    dt = pd.to_datetime(dt_str, utc=True)

    # Convert to EST
    est_timezone = pytz.timezone('America/New_York')
    return dt.tz_convert(est_timezone)

# Function to handle the conversion of actual times based on flight status
def handle_flight_status(row):
    if row['flight_status'] == 'landed':
        # Convert both departure and arrival times
        row['departure.actual.EST'] = convert_to_est(row['departure.actual'])
        row['arrival.actual.EST'] = convert_to_est(row['arrival.actual'])
    elif row['flight_status'] == 'active':
        # Convert only departure time
        row['departure.actual.EST'] = convert_to_est(row['departure.actual'])
        row['arrival.actual.EST'] = None  # Not applicable for active flights
    else:  # For 'scheduled' or any other status
        # Do not convert either
        row['departure.actual.EST'] = None
        row['arrival.actual.EST'] = None

    return row

# Function to add EST conversion for scheduled and estimated times
def add_est_columns(df):
    est_columns = {
        'departure.scheduled.EST': 'departure.scheduled',
        'departure.estimated.EST': 'departure.estimated',
        'arrival.scheduled.EST': 'arrival.scheduled',
        'arrival.estimated.EST': 'arrival.estimated'
    }

    for new_col, time_col in est_columns.items():
        df[new_col] = df[time_col].apply(convert_to_est)

    return df

# Function to apply all EST conversions
def apply_all_est_conversions(df):
    # Convert scheduled and estimated times
    df = add_est_columns(df)
    # Convert actual times based on flight status
    df = df.apply(handle_flight_status, axis=1)

    return df[['flight.number','flight_date','airline.name','flight_status','departure.airport','departure.timezone',
              'arrival.airport','arrival.timezone','departure.scheduled','departure.estimated','departure.actual','arrival.scheduled',
              'arrival.estimated','arrival.actual','departure.scheduled.EST','departure.estimated.EST','departure.actual.EST','arrival.scheduled.EST',
              'arrival.estimated.EST','arrival.actual.EST']]

# Assuming 'df' is your DataFrame
df = apply_all_est_conversions(df)
df.head()
```
